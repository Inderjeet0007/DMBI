# Data Visualization

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("data_university.csv")
df.head(10)
```

```python
df['national_rank'].hist(bins=15)
plt.show()
```

```python
df['quality_of_education'].hist(bins=20)
plt.show()
```

```python
#to resize the figure
plt.rcParams["figure.figsize"] =(14,8)
plt.hist(df['country'],bins=20)
plt.xlabel('country')
plt.show()
```

```python
plt.bar(df["country"],df["quality_of_education"],label="Bar Chart")
plt.xlabel('country')
plt.ylabel('quality_of_education')
plt.show()
```

```python
plt.bar(df["country"],df["world_rank"],label="Bar Chart")
plt.xlabel('country')
plt.ylabel('world')
plt.show()
```

USA has maximum number of universities

```python
plt.rcParams["figure.figsize"] =(14,8)
plt.bar(df["country"],df["national_rank"],label="Bar Chart")
plt.xlabel('country')
plt.ylabel('national_rank')
plt.show()
```

# Box Plot

```python
plt.rcParams["figure.figsize"] =(5,5)
df.boxplot(column='national_rank')
plt.show()
```

```python
plt.rcParams["figure.figsize"] =(14,8)
df.boxplot(column='national_rank', by="country")
plt.show()
```

Analysis 

```python
df.boxplot(column='national_rank', by="world_rank")
plt.show()
```

National rank directly proportional to World rank

```python
plt.rcParams["figure.figsize"] =(14,8)
sns.boxplot(x="national_rank", y="world_rank", data=df, palette="PRGn")
```
